```python
from matplotlib import rcParams
rcParams['font.cursive']
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d
#import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datos1 = pd.read_csv('caso1.txt', sep = '\s+')
datos2 = pd.read_csv('caso2.txt', sep = '\s+')
datos3 = pd.read_csv('caso3.txt', sep = '\s+')
datos4 = pd.read_csv('caso4.txt', sep = '\s+')

phasespace = pd.read_csv('espaciofase.txt', sep = '\s+')

taups = pd.DataFrame(phasespace['A']).values.flatten()
taukn = pd.DataFrame(phasespace['C']).values.flatten()

sr1 = pd.DataFrame(datos1['Sr']).values.flatten()
s = np.sqrt(sr1)
```

```python
# Creating figure
fig = plt.figure(figsize=(10,7))
plt.axvline(1.331,c = 'black', ls ='--')
plt.axvline(1.437,c ='black', ls ='--')
plt.ylabel(r'$\sigma_{\bar{K}N - \bar{K}N}$',fontsize=15)
plt.xlabel(r'$\sqrt{s}$ [Mev]',fontsize=15)
plt.xlim(1.3,1.5)

# Creating plot
plt.plot(s,taups, label = 'PS')
plt.plot(s,taukn, label = 'KN')
plt.text(1.325, 0.18, r'$\pi \Sigma$',fontsize=15)
plt.text(1.437, 0.18, r'$\bar{K}$N',fontsize=15)
plt.legend()

plt.show
```

```python
rps1 = pd.DataFrame(datos1['Rpps']).values.flatten()
rkn1 = pd.DataFrame(datos1['Rpkn']).values.flatten()
rpk1 = pd.DataFrame(datos1['Rppk']).values.flatten()
ips1 = pd.DataFrame(datos1['Ipps']).values.flatten()
ikn1 = pd.DataFrame(datos1['Ipkn']).values.flatten()
ipk1 = pd.DataFrame(datos1['Ippk']).values.flatten()

realps1 = rps1*taups
realkn1 = rkn1*taukn
realpk1 = rpk1*taukn
Imagps1 = ips1*taups
Imagkn1 = ikn1*taukn
Imagpk1 = ipk1*taukn
```
